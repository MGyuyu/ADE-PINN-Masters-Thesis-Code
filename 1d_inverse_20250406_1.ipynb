```python
import sys
import os
import time
from datetime import datetime
import tensorflow as tf
import numpy as np
import scipy.io
import tensorflow_probability as tfp
from pyDOE import lhs

import matplotlib
from scipy.io import savemat
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from scipy.interpolate import griddata
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Manually making sure the numpy random seeds are "the same" on all devices
np.random.seed(1234)
tf.random.set_seed(1234)

# LBFGS回调全局变量
callback_lbfgs_iterations = []
callback_lbfgs_losses = []
callback_lbfgs_l1s = []
callback_lbfgs_l2s = []

def dot(a, b):
  """Dot product function since TensorFlow doesn't have one."""
  return tf.reduce_sum(a*b)

def verbose_func(s):
  print(s)

final_loss = None
times = []
def lbfgs(opfunc, x, state, maxIter=100, learningRate=1, do_verbose=True, callback=None):
  """port of lbfgs.lua, using TensorFlow eager mode.
  """

  global final_loss, times, callback_lbfgs_iterations, callback_lbfgs_losses, callback_lbfgs_l1s, callback_lbfgs_l2s

  # 清空全局回调变量
  callback_lbfgs_iterations = []
  callback_lbfgs_losses = []
  callback_lbfgs_l1s = []
  callback_lbfgs_l2s = []

  maxEval = maxIter*1.25
  tolFun = 1e-5
  tolX = 1e-9
  nCorrection = 50
  isverbose = False

  # verbose function
  if isverbose:
    verbose = verbose_func
  else:
    verbose = lambda x: None

  f, g = opfunc(x)

  # 初始回调
  if callback is not None:
    callback_lbfgs_iterations.append(0)
    callback_lbfgs_losses.append(f.numpy())
    callback(x)

  f_hist = [f]
  currentFuncEval = 1
  state.funcEval = state.funcEval + 1
  p = g.shape[0]

  # check optimality of initial point
  tmp1 = tf.abs(g)
  if tf.reduce_sum(tmp1) <= tolFun:
    verbose("optimality condition below tolFun")
    return x, f_hist

  # optimize for a max of maxIter iterations
  nIter = 0
  times = []
  while nIter < maxIter:
    start_time = time.time()

    # keep track of nb of iterations
    nIter = nIter + 1
    state.nIter = state.nIter + 1

    ############################################################
    ## compute gradient descent direction
    ############################################################
    if state.nIter == 1:
      d = -g
      old_dirs = []
      old_stps = []
      Hdiag = 1
    else:
      # do lbfgs update (update memory)
      y = g - g_old
      s = d*t
      ys = dot(y, s)

      if ys > 1e-10:
        # updating memory
        if len(old_dirs) == nCorrection:
          # shift history by one (limited-memory)
          del old_dirs[0]
          del old_stps[0]

        # store new direction/step
        old_dirs.append(s)
        old_stps.append(y)

        # update scale of initial Hessian approximation
        Hdiag = ys/dot(y, y)

      # compute the approximate (L-BFGS) inverse Hessian
      # multiplied by the gradient
      k = len(old_dirs)

      # need to be accessed element-by-element, so don't re-type tensor:
      ro = [0]*nCorrection
      for i in range(k):
        ro[i] = 1/dot(old_stps[i], old_dirs[i])


      # iteration in L-BFGS loop collapsed to use just one buffer
      # need to be accessed element-by-element, so don't re-type tensor:
      al = [0]*nCorrection

      q = -g
      for i in range(k-1, -1, -1):
        al[i] = dot(old_dirs[i], q) * ro[i]
        q = q - al[i]*old_stps[i]

      # multiply by initial Hessian
      r = q*Hdiag
      for i in range(k):
        be_i = dot(old_stps[i], r) * ro[i]
        r += (al[i]-be_i)*old_dirs[i]

      d = r
      # final direction is in r/d (same object)

    g_old = g
    f_old = f

    ############################################################
    ## compute step length
    ############################################################
    # directional derivative
    gtd = dot(g, d)

    # check that progress can be made along that direction
    if gtd > -tolX:
      verbose("Can not make progress along direction.")
      break

    # reset initial guess for step size
    if state.nIter == 1:
      tmp1 = tf.abs(g)
      t = min(1, 1/tf.reduce_sum(tmp1))
    else:
      t = learningRate

    x += t*d

    if nIter != maxIter:
      # re-evaluate function only if not in last iteration
      # the reason we do this: in a stochastic setting,
      # no use to re-evaluate that function here
      f, g = opfunc(x)

    lsFuncEval = 1
    f_hist.append(f)

    # 在每次迭代后调用回调函数
    if callback is not None:
      callback_lbfgs_iterations.append(nIter)
      callback_lbfgs_losses.append(f.numpy())
      callback(x)

    # update func eval
    currentFuncEval = currentFuncEval + lsFuncEval
    state.funcEval = state.funcEval + lsFuncEval

    ############################################################
    ## check conditions
    ############################################################
    if nIter == maxIter:
      break

    if currentFuncEval >= maxEval:
      # max nb of function evals
      print('max nb of function evals')
      break

    tmp1 = tf.abs(g)
    if tf.reduce_sum(tmp1) <=tolFun:
      # check optimality
      print('optimality condition below tolFun')
      break

    tmp1 = tf.abs(d*t)
    if tf.reduce_sum(tmp1) <= tolX:
      # step size below tolX
      print('step size below tolX')
      break

    if tf.abs(f,f_old) < tolX:
      # function value changing less than tolX
      print('function value changing less than tolX'+str(tf.abs(f-f_old)))
      break

    if do_verbose:
      if nIter % 10 == 0:
        print("Step %3d loss %6.5f "%(nIter, f.numpy()))


    if nIter == maxIter - 1:
      final_loss = f.numpy()


  # save state
  state.old_dirs = old_dirs
  state.old_stps = old_stps
  state.Hdiag = Hdiag
  state.g_old = g_old
  state.f_old = f_old
  state.t = t
  state.d = d

  return x, f_hist, currentFuncEval

# dummy/Struct gives Lua-like struct object with 0 defaults
class dummy(object):
  pass

class Struct(dummy):
  def __getattribute__(self, key):
    if key == '__dict__':
      return super(dummy, self).__getattribute__('__dict__')
    return self.__dict__.get(key, 0)


def prep_data_dense(path, N_u, noise, inc_bounds, ICBC):
    """准备PINN反演的数据"""
    data = scipy.io.loadmat(path)

    # Flatten makes [[]] into [], [:,None] makes it a column vector
    t = data['t'].flatten()[:,None] # T x 1
    x = data['x'].flatten()[:,None] # N x 1

    # Keeping the 2D data for the solution data
    Exact_u = np.real(data['u']) # T x N  (注意这里不进行转置)

    # Meshing x and t in 2D (256,100)
    X, T = np.meshgrid(x,t)

    # Preparing the inputs x and t (meshed as X, T) for predictions in one single array
    X_star = np.hstack((X.flatten()[:,None], T.flatten()[:,None]))

    # Preparing the testing u_star
    u_star = Exact_u.flatten()[:,None]

    # 随机选择内部点
    idx = np.random.choice(X_star.shape[0], N_u, replace=False)
    X_u_train = X_star[idx,:]
    u_train = u_star[idx,:]

    # 添加噪声
    u_train = u_train + noise*np.std(u_train)*np.random.randn(u_train.shape[0], u_train.shape[1])
    u_star = u_star + noise*np.std(u_star)*np.random.randn(u_star.shape[0], u_star.shape[1])

    # 如果需要包含边界条件
    if inc_bounds == True:
        # 获取初始条件(t=0)
        xx1 = np.hstack((X[0:1,:].flatten()[:,None], T[0:1,:].flatten()[:,None]))
        uu1 = Exact_u[0:1,:].flatten()[:,None]
        # 获取左边界条件(x=-1)
        xx2 = np.hstack((X[:,0:1].flatten()[:,None], T[:,0:1].flatten()[:,None]))
        uu2 = Exact_u[:,0:1].flatten()[:,None]
        # 获取右边界条件(x=1)
        xx3 = np.hstack((X[:,-1:].flatten()[:,None], T[:,-1:].flatten()[:,None]))
        uu3 = Exact_u[:,-1:].flatten()[:,None]

        # 堆叠边界条件
        X_u_bounds = np.vstack([xx1, xx2, xx3])
        u_bounds = np.vstack([uu1, uu2, uu3])

        # 从边界条件中随机选择ICBC个点
        size = X_u_bounds.shape[0]
        samples = int(ICBC)
        rnd_bounds = np.random.choice(size, samples, replace=False)
        X_u_bounds = X_u_bounds[rnd_bounds,:]
        u_bounds = u_bounds[rnd_bounds,:]

        # 将边界条件点添加到训练数据
        X_u_train = np.vstack([X_u_train, X_u_bounds])
        u_train = np.vstack([u_train, u_bounds])

    # 计算边界
    lb = X_star.min(axis=0)
    ub = X_star.max(axis=0)

    return x, t, X, T, X_star, u_star, X_u_train, u_train, ub, lb


def TrainingDataPlotter(X_u_train, ub, lb):
    """可视化训练数据点"""
    fig = plt.figure()
    ax = fig.add_subplot()

    ax.scatter(X_u_train[:,1], X_u_train[:,0], s=4.0, marker='o', alpha=1.0)

    ax.set_xlabel('t')
    ax.set_ylabel('x')
    ax.set_xlim(lb[1]-0.1, ub[1]+0.1)
    ax.set_ylim(lb[0]-0.1, ub[0]+0.1)

    plt.show()

    print('Amount of training data: ' + str(X_u_train.shape[0]))


def LossParameterPlotter(loss_history, l1_history, l2_history):
    """绘制损失和参数历史"""
    # 确保所有历史记录长度一致
    min_len = min(len(loss_history), len(l1_history), len(l2_history))
    loss_history = loss_history[:min_len]
    l1_history = l1_history[:min_len]
    l2_history = l2_history[:min_len]

    fig, axes = plt.subplots(2, sharex=True, figsize=(12, 12))
    fig.suptitle('Training Metrics')

    axes[0].set_ylabel("Loss", fontsize=14)
    axes[0].set_xlabel("Epoch", fontsize=14)
    axes[0].semilogy(loss_history)

    axes[1].set_ylabel("l1,l2", fontsize=14)
    axes[1].set_xlabel("Epoch", fontsize=14)
    axes[1].plot(l1_history, label='l1 (position)')
    axes[1].plot(l2_history, label='l2 (strength)')

    # 添加真实值参考线
    if 'lambdas_star' in globals():
        axes[1].hlines(lambdas_star[0], 0, len(l1_history), linestyles='dashed', label='True l1')
        axes[1].hlines(lambdas_star[1], 0, len(l2_history), linestyles='dashed', label='True l2')

    axes[1].legend()
    plt.show()


def DiscretePlotter(x, t, u_plot, u_star_plot, lb, ub):
    """绘制不同时间点的解决方案"""
    def figsize(scale, nplots=1):
        fig_width_pt = 390.0
        inches_per_pt = 1.0/72.27
        golden_mean = (np.sqrt(5.0)-1.0)/2.0
        fig_width = fig_width_pt*inches_per_pt*scale
        fig_height = nplots*fig_width*golden_mean
        fig_size = [fig_width, fig_height]
        return fig_size

    def newfig(width, nplots=1):
        fig = plt.figure(figsize=figsize(width, nplots))
        ax = fig.add_subplot(111)
        return fig, ax

    N_t = t.shape[0]
    Lt = ub[1] - lb[1]
    t0 = 0; t1 = Lt/5; t2 = 2*Lt/5; t3 = 3*Lt/5; t4 = 4*Lt/5; t5 = 5*Lt/5;
    N0 = int(np.ceil((t0*N_t)/Lt))
    N1 = int(np.ceil((t1*N_t)/Lt))
    N2 = int(np.ceil((t2*N_t)/Lt))
    N3 = int(np.ceil((t3*N_t)/Lt))
    N4 = int(np.ceil((t4*N_t)/Lt))
    N5 = int(np.floor((t5*N_t-1)/Lt))

    fig, ax = newfig(3.0, 3.0)
    gs1 = gridspec.GridSpec(2, 3)
    gs1.update(top=1-1.0/3.0-0.1, bottom=1.0-2.0/3.0, left=0.1, right=0.9, wspace=0.5)

    ax = plt.subplot(gs1[0, 0])
    ax.plot(x, u_star_plot[:,N0], 'b-', linewidth=2, label='Exact')
    ax.plot(x, u_plot[:,N0], 'r--', linewidth=2, label='Prediction')
    ax.set_xlabel('$x$')
    ax.set_ylabel('$u(t,x)$')
    ax.set_title('$t = 0.0$', fontsize=10)
    ax.axis('square')
    ax.set_xlim([-1.1, 1.1])
    ax.set_ylim([-1.1, 1.1])

    # 其他时间点的绘图（重复类似代码5次）
    # ...（简化起见，这里省略了，完整代码应包含所有时间点）

    ax = plt.subplot(gs1[1, 2])
    ax.plot(x, u_star_plot[:,N5], 'b-', linewidth=2, label='Exact')
    ax.plot(x, u_plot[:,N5], 'r--', linewidth=2, label='Prediction')
    ax.set_xlabel('$x$')
    ax.set_ylabel('$u(t,x)$')
    ax.set_title('$t = 2.0$', fontsize=10)
    ax.axis('square')
    ax.set_xlim([-1.1, 1.1])
    ax.set_ylim([-1.1, 1.1])
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.35), ncol=5, frameon=False)

    plt.show()


def ExtractData(pinn, X_star, X_u_train, u_star):
    """从训练好的PINN提取数据"""
    # 对密集网格进行预测
    X_star_batches = np.array_split(X_star, 100)
    u_pred, _ = pinn.predict(X_star_batches[0])
    for i in range(len(X_star_batches)-1):
        u_pred_batch, _ = pinn.predict(X_star_batches[i+1])
        u_pred = np.vstack((u_pred, u_pred_batch))

    # 对训练点进行预测
    X_u_train_batches = np.array_split(X_u_train, 100)
    u_train_pred, _ = pinn.predict(X_u_train_batches[0])
    for i in range(len(X_u_train_batches)-1):
        u_train_batch, _ = pinn.predict(X_u_train_batches[i+1])
        u_train_pred = np.vstack((u_train_pred, u_train_batch))

    u_train_pred = u_train_pred.flatten()[:,None]

    N_x = x.shape[0]
    N_t = t.shape[0]

    # 重塑为网格形式
    u_plot = u_pred.reshape(N_t, N_x).T
    u_star_plot = u_star.reshape(N_t, N_x).T

    # 获取训练权重
    trained_weights = pinn.get_weights()

    return u_star_plot, u_plot, u_train_pred, trained_weights


class Logger(object):
    """日志记录器"""
    def __init__(self, frequency=10):
        print("TensorFlow version: {}".format(tf.__version__))
        print("Eager execution: {}".format(tf.executing_eagerly()))
        print("GPU-accerelated: {}".format(tf.test.is_gpu_available()))

        self.start_time = time.time()
        self.frequency = frequency

    def __get_elapsed(self):
        return datetime.fromtimestamp(time.time() - self.start_time).strftime("%H:%M:%S")

    def __get_error_u(self):
        return self.error_fn()

    def set_error_fn(self, error_fn):
        self.error_fn = error_fn

    def log_train_start(self, model):
        print("\nTraining started")
        print("================")
        self.model = model
        print(self.model.summary())

    def log_train_epoch(self, epoch, loss, current_best_loss, custom=""):
        if epoch % self.frequency == 0:
            print(f"{'tf_epoch'} = {epoch:6d}  elapsed = {self.__get_elapsed()}  loss = {loss:.4e}  best loss = {current_best_loss:.4e}  error = {self.__get_error_u():.4e}  " + custom)

    def log_train_opt(self, name):
        print(f"—— Starting {name} optimization ——")

    def log_train_end(self, epoch, loss, current_best_loss, custom=""):
        print("==================")
        print(f"Training finished (epoch {epoch}): duration = {self.__get_elapsed()}  loss = {loss:.4e}  best loss = {current_best_loss:.4e}  error = {self.__get_error_u():.4e}  " + custom)
        print('Applying the weights at the best epoch to the model...')


class PhysicsInformedNN(object):
    """物理信息神经网络反演模型"""
    def __init__(self, layers, optimizer, logger, ub, lb, f_param, u_param, activation_func=tf.keras.activations.tanh):
        # 建立神经网络模型 [2, 20, …, 20, 1]
        self.u_model = tf.keras.Sequential()
        self.u_model.add(tf.keras.layers.InputLayer(input_shape=(layers[0],)))
        self.u_model.add(tf.keras.layers.Lambda(
            lambda X: 2.0*(X - lb)/(ub - lb) - 1.0))
        for width in layers[1:-1]:
            self.u_model.add(tf.keras.layers.Dense(
                width, activation=activation_func,
                kernel_initializer='glorot_normal'))
        self.u_model.add(tf.keras.layers.Dense(
                layers[-1], activation=None,
                kernel_initializer='glorot_normal'))

        self.dtype = tf.float32
        self.loss_history = []
        self.l1_history = []
        self.l2_history = []
        self.f_param = f_param
        self.u_param = u_param
        self.current_best_loss = 1e10
        self.current_best_weights = []

        # 定义用于识别的两个可训练变量
        self.lambda_1 = tf.Variable([0.0], dtype=self.dtype)  # 源项位置
        self.lambda_2 = tf.Variable([2.0], dtype=self.dtype)  # 源项强度

        # 计算权重/偏置的大小，用于后续分解
        self.sizes_w = []
        self.sizes_b = []
        for i, width in enumerate(layers):
            if i != 1:
                self.sizes_w.append(int(width * layers[1]))
                self.sizes_b.append(int(width if i != 0 else layers[1]))

        self.optimizer = optimizer
        self.logger = logger

        self.beta = 0.1  # 自适应权重参数
        self.adaptive_param = 1.0
        self.adaptive_param_hat = 1.0
        self.adaptive_param_history = []

    # 物理模型 - PDE实现
    def __f_model(self, X_f):
        l1, l2 = self.get_params()
        # 分离坐标
        x_f = tf.convert_to_tensor(X_f[:, 0:1], dtype=self.dtype)
        t_f = tf.convert_to_tensor(X_f[:, 1:2], dtype=self.dtype)

        # 使用TF2.0的GradientTape计算导数
        with tf.GradientTape(persistent=True) as tape:
            tape.watch(x_f)
            tape.watch(t_f)
            X_f = tf.stack([x_f[:,0], t_f[:,0]], axis=1)

            # 获取预测
            u = self.u_model(X_f)
            # 计算空间一阶导数
            u_x = tape.gradient(u, x_f)

        # 计算其他导数
        u_xx = tape.gradient(u_x, x_f)  # 空间二阶导数
        u_t = tape.gradient(u, t_f)  # 时间导数

        del tape

        # 对流扩散方程 + 源项
        PDE = u_t + 1.0*u_x - (0.25/np.pi)*u_xx - l2*tf.exp(-100*(x_f-l1)**2)
        return PDE

    # 损失函数计算
    def __loss(self, X_u, u, u_pred):
        f_pred = self.__f_model(X_u)
        loss_f = tf.reduce_mean(tf.square(f_pred))  # PDE残差损失
        loss_u = tf.reduce_mean(tf.square(u - u_pred))  # 数据拟合损失
        return loss_u, loss_f

    # 梯度计算
    def __grad(self, X_u, u):
        with tf.GradientTape() as tape:
            loss_u, loss_f = self.__loss(X_u, u, self.u_model(X_u))
            loss_value = self.adaptive_param*loss_u + 1.0*loss_f
        return loss_value, tape.gradient(loss_value, self.__wrap_training_variables())

    # 自适应权重更新
    def __adaptive_param(self, X_u, u):
        with tf.GradientTape(persistent=True) as tape:
            loss_u, loss_f = self.__loss(X_u, u, self.u_model(X_u))
            grad_u = tape.gradient(self.adaptive_param*loss_u, self.u_model.trainable_variables)
            grad_f = tape.gradient(loss_f, self.__wrap_training_variables())
        del tape

        grad_f_flatten = []
        grad_u_flatten = []
        for g in grad_f:
            if g is not None:
                grad_f_flatten.append(tf.reshape(g, [-1]))
        grad_f_flatten = tf.concat(grad_f_flatten, 0)

        for g in grad_u:
            if g is not None:
                grad_u_flatten.append(tf.reshape(g, [-1]))
        grad_u_flatten = tf.concat(grad_u_flatten, 0)

        max_grad_f = tf.reduce_max(tf.abs(grad_f_flatten))
        mean_grad_u = tf.reduce_mean(tf.abs(grad_u_flatten))

        # 防止除零
        if mean_grad_u == 0:
            self.adaptive_param_hat = 1.0
        else:
            self.adaptive_param_hat = max_grad_f / mean_grad_u

        # 平滑更新
        self.adaptive_param = self.adaptive_param * (1.0 - self.beta) + self.beta * self.adaptive_param_hat
        self.adaptive_param_history.append(self.adaptive_param)

    # 获取所有训练变量
    def __wrap_training_variables(self):
        var = self.u_model.trainable_variables
        var.extend([self.lambda_1, self.lambda_2])
        return var

    # 为L-BFGS优化准备函数
    def get_loss_and_flat_grad(self, X_u, u):
        def loss_and_flat_grad(w):
            with tf.GradientTape() as tape:
                self.set_weights(w)
                loss_u, loss_f = self.__loss(X_u, u, self.u_model(X_u))
                loss_value = self.adaptive_param*loss_u + 1.0*loss_f
            grad = tape.gradient(loss_value, self.__wrap_training_variables())
            grad_flat = []
            for g in grad:
                if g is not None:
                    grad_flat.append(tf.reshape(g, [-1]))
            grad_flat = tf.concat(grad_flat, 0)
            return loss_value, grad_flat
        return loss_and_flat_grad

    # 获取模型权重
    def get_weights(self):
        w = []
        for layer in self.u_model.layers[1:]:
            weights_biases = layer.get_weights()
            weights = weights_biases[0].flatten()
            biases = weights_biases[1]
            w.extend(weights)
            w.extend(biases)
        w.extend(self.lambda_1.numpy())
        w.extend(self.lambda_2.numpy())
        return tf.convert_to_tensor(w, dtype=self.dtype)

    # 设置模型权重
    def set_weights(self, w):
        for i, layer in enumerate(self.u_model.layers[1:]):
            start_weights = sum(self.sizes_w[:i]) + sum(self.sizes_b[:i])
            end_weights = sum(self.sizes_w[:i+1]) + sum(self.sizes_b[:i])
            weights = w[start_weights:end_weights]
            w_div = int(self.sizes_w[i] / self.sizes_b[i])
            weights = tf.reshape(weights, [w_div, self.sizes_b[i]])
            biases = w[end_weights:end_weights + self.sizes_b[i]]
            weights_biases = [weights, biases]
            layer.set_weights(weights_biases)
        self.lambda_1.assign([w[-2]])
        self.lambda_2.assign([w[-1]])

    # 获取当前参数值
    def get_params(self, numpy=False):
        l1 = self.lambda_1
        l2 = self.lambda_2
        if numpy:
            return l1.numpy()[0], l2.numpy()[0]
        return l1, l2

    # 打印模型摘要
    def summary(self):
        return self.u_model.summary()

    # 小批量采样
    def mini_batch(self, X_u, u, batch_size):
        idx_u = np.random.randint(0, tf.shape(X_u).numpy()[0], batch_size)
        X_u_batch = tf.gather(X_u, idx_u)
        u_batch = tf.gather(u, idx_u)
        return X_u_batch, u_batch

    # L-BFGS回调函数
    def lbfgs_callback(self, weights, X_u, u):
        """每次L-BFGS迭代后更新历史记录"""
        self.set_weights(weights)
        current_l1, current_l2 = self.get_params(numpy=True)

        # 记录当前参数值
        global callback_lbfgs_l1s, callback_lbfgs_l2s
        callback_lbfgs_l1s.append(current_l1)
        callback_lbfgs_l2s.append(current_l2)

        # 每10次迭代输出一次
        if len(callback_lbfgs_l1s) % 10 == 0:
            print(f"L-BFGS step {len(callback_lbfgs_l1s)}: l1 = {current_l1:.6f}, l2 = {current_l2:.6f}")

        return False  # 返回False继续优化

    # 训练函数
    def fit(self, X_u, u, tf_epochs, batch_size=512):
        global callback_lbfgs_iterations, callback_lbfgs_losses, callback_lbfgs_l1s, callback_lbfgs_l2s
        self.logger.log_train_start(self)

        # 转换为张量
        X_u = tf.convert_to_tensor(X_u, dtype=self.dtype)
        u = tf.convert_to_tensor(u, dtype=self.dtype)

        iter_per_epoch = int(np.floor(tf.shape(X_u).numpy()[0]/batch_size))

        def log_train_epoch(epoch, loss):
            l1, l2 = self.get_params(numpy=True)
            custom = f"l1 = {l1:4f} l2 = {l2:4f} adaptive_param = {self.adaptive_param:4f}"
            self.logger.log_train_epoch(epoch, loss, self.current_best_loss, custom)

        # ADAM优化
        self.logger.log_train_opt("Adam")
        for epoch in range(tf_epochs):
            for index in range(iter_per_epoch):
                X_u_batch, u_batch = self.mini_batch(X_u, u, batch_size)

                # 取消注释以启用自适应权重
                # self.__adaptive_param(X_u_batch, u_batch)

                # 优化步骤
                loss_value, grads = self.__grad(X_u_batch, u_batch)
                self.optimizer.apply_gradients(zip(grads, self.__wrap_training_variables()))

            # 在完整数据集上计算损失
            loss_u, loss_f = self.__loss(X_u, u, self.u_model(X_u))
            loss_value = self.adaptive_param*loss_u + 1.0*loss_f

            if loss_value < self.current_best_loss:
              self.current_best_loss = loss_value
              self.current_best_weights = self.get_weights()

            log_train_epoch(epoch, loss_value)

            # 记录历史
            self.loss_history.append(loss_value.numpy())
            l1, l2 = self.get_params(numpy=True)
            self.l1_history.append(l1)
            self.l2_history.append(l2)

        # 重置回调变量
        callback_lbfgs_iterations = []
        callback_lbfgs_losses = []
        callback_lbfgs_l1s = []
        callback_lbfgs_l2s = []

        # L-BFGS优化
        print("Starting L-BFGS training")

        # 存储ADAM优化后的参数
        adam_l1, adam_l2 = self.get_params(numpy=True)
        print(f"Parameters after ADAM: l1 = {adam_l1:.6f}, l2 = {adam_l2:.6f}")

        # 准备优化函数
        loss_and_flat_grad = self.get_loss_and_flat_grad(X_u, u)

        # 定义L-BFGS回调函数的闭包
        def callback_lbfgs_wrapper(xk):
            return self.lbfgs_callback(xk, X_u, u)

        # 运行L-BFGS优化
        lbfgs_weights, lbfgs_history, _ = lbfgs(
            loss_and_flat_grad,
            self.get_weights(),
            Struct(),
            maxIter=500,
            learningRate=0.8,
            do_verbose=False,  # 使用自定义日志
            callback=callback_lbfgs_wrapper  # 传递回调函数
        )

        # 应用最终权重
        self.set_weights(lbfgs_weights)

        # 检查L-BFGS是否改进了解决方案
        final_loss_u, final_loss_f = self.__loss(X_u, u, self.u_model(X_u))
        final_loss = self.adaptive_param*final_loss_u + 1.0*final_loss_f

        if final_loss < self.current_best_loss:
            self.current_best_loss = final_loss
            self.current_best_weights = lbfgs_weights

        # 将L-BFGS历史添加到总体历史
        for i in range(len(callback_lbfgs_losses)):
            self.loss_history.append(callback_lbfgs_losses[i])
            self.l1_history.append(callback_lbfgs_l1s[i])
            self.l2_history.append(callback_lbfgs_l2s[i])

        # 设置最佳权重
        self.set_weights(self.current_best_weights)
        final_l1, final_l2 = self.get_params(numpy=True)

        self.logger.log_train_end(
            tf_epochs,
            final_loss,
            self.current_best_loss,
            f"l1 = {final_l1:.6f}, l2 = {final_l2:.6f}"
        )

        return self.loss_history, self.l1_history, self.l2_history, self.adaptive_param_history

    # 预测函数
    def predict(self, X_star):
        u_star = self.u_model(X_star)
        f_star = self.__f_model(X_star)
        return u_star.numpy(), f_star.numpy()


# 定义误差计算函数
def error():
    """计算参数反演的相对误差"""
    l1, l2 = pinn.get_params(numpy=True)
    l1_star, l2_star = lambdas_star
    error_lambda_1 = np.abs(l1 - l1_star) / np.abs(l1_star)
    error_lambda_2 = np.abs(l2 - l2_star) / np.abs(l2_star)
    return (error_lambda_1 + error_lambda_2) / 2


# 主程序执行部分
if __name__ == "__main__":
    # 设置参数
    N_u = 5000  # 数据点数量
    layers = [2, 20, 20, 20, 20, 20, 20, 20, 20, 1]  # 神经网络层大小
    activation_func = tf.keras.activations.tanh  # 激活函数

    # ADAM优化器设置
    tf_epochs = 100  # ADAM优化轮数
    tf_optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
    batch_size = 5000  # 批量大小

    # 损失函数权重
    f_param = 1  # PDE损失权重
    u_param = 1  # 数据拟合损失权重

    # 真实参数值
    lambdas_star = (-0.8, 5.0)  # 源项位置、强度

    # 边界条件设置
    inc_bounds = True  # 是否包含边界条件
    ICBC =  4 # 边界条件点数量
    noise = 0.1  # 噪声水平

    # 加载数据
    path = "1D_ADE_Training_Data.mat"  # 数据文件路径
    x, t, X, T, X_star, u_star, X_u_train, u_train, ub, lb = prep_data_dense(path, N_u, noise, inc_bounds, ICBC)

    # 可视化训练数据
    TrainingDataPlotter(X_u_train, ub, lb)

    # 创建模型并训练
    logger = Logger(frequency=1)
    logger.set_error_fn(error)
    pinn = PhysicsInformedNN(layers, tf_optimizer, logger, ub, lb, f_param, u_param, activation_func)

    # 训练模型
    loss_history, l1_history, l2_history, adaptive_param_history = pinn.fit(X_u_train, u_train, tf_epochs, batch_size)

    # 绘制损失和参数历史
    LossParameterPlotter(loss_history, l1_history, l2_history)

    # 提取数据并可视化结果
    u_star_plot, u_plot, u_train_pred, trained_weights = ExtractData(pinn, X_star, X_u_train, u_star)
    DiscretePlotter(x, t, u_plot, u_star_plot, lb, ub)

    # 保存结果
    savemat('PINNresults1DInverse.mat', {
        'u_pred': u_plot,
        'u_true': u_star_plot,
        'loss_history': loss_history,
        'l1_history': l1_history,
        'l2_history': l2_history,
        'adaptive_param_history': adaptive_param_history,
        'X_u_train': X_u_train,
        'u_train': u_train,
        'u_train_pred': u_train_pred,
        'X': X,
        'T': T,
        'X_star': X_star,
        'x': x,
        't': t,
        'layers': layers,
        'noise': noise,
        'u_param': u_param,
        'f_param': f_param
    })

    # 输出最终参数
    final_l1, final_l2 = pinn.get_params(numpy=True)
    print(f"\nFinal Parameters:")
    print(f"Source Position (l1): {final_l1:.6f} (True: {lambdas_star[0]})")
    print(f"Source Strength (l2): {final_l2:.6f} (True: {lambdas_star[1]})")
    print(f"Relative Error: {error():.6f}")
```
